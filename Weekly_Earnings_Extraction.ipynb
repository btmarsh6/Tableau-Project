```python
import pandas as pd
import json
from pprint import pprint
```

```python
with open('data/weekly_earnings.json') as f:
    nested_json = json.load(f)
pprint(nested_json)
```

```python
data = nested_json['data']
data
```

```python
columns = [ x['name'] for x in nested_json['meta']['view']['columns']]
```

```python
columns
```

```python
weekly_earnings_df = pd.DataFrame(data, columns=columns)
weekly_earnings_df
```

```python
weekly_earnings_df.to_csv('data/weekly_earnings.csv')
```
